## Juego Trivial

Base de datos con las preguntas

```python
preguntas = {
    1:{ "pregunta": "Un señor bajito con un anillo",
        "opciones": "(A) Harry Potter, (B) El señor de los anillos y (C) Los vengadores",
        "respuesta_correcta": "B" },
    2:{ "pregunta": "El número once se pelea con un tio feo",
        "opciones": "(A) Stranger things, (B) El señor de los anillos y (C) Los vengadores",
        "respuesta_correcta": "A"}
}
```

Importamos libreria random y creamos un número aleatorio

```python
import random

n = random.randint(1,2)
```

Iniciamos el juego

```python
print("Bienvenido al juego")

while True:
    
    print(preguntas[n]["pregunta"])
    print(preguntas[n]["opciones"])

    respuesta = input("Introduce tu respuesta").upper()

    if respuesta == preguntas[n]["respuesta_correcta"]:
        print("Es correcto")
    else:
        print("Prueba otra vez")
        
    n = random.randint(1,2)
```
