```python
nombre=input("Nombre del jugador").title().strip()
while nombre == " " or len(nombre) == 0:
    print("Introduce un nombre:")
    nombre=input("Vuelve a introducir tu nombre:")

print(f"¡Bienvenida al juego, {nombre}!")


while True:
    try:
        num_preguntas = int(input("¿Cuántas preguntas quieres hacer? Elige al menos 5"))
        
        if num_preguntas < 5 or num_preguntas > 10:
            print("Por favor, introduce un número entre el 5 y 10.")
        else:
            break

    except ValueError:
        print("Entrada no válida. Tienes que introducir un número del 5 al 10.")

print("¡Que comience el juego!")
```

```python
import random
mazo= random.sample(preguntas, num_preguntas)
```

```python
contador=1
puntuacion=0
vidas=3

for i in mazo:
    print(f"Pregunta {contador}: {i['pregunta']}")
    for letra, opcion in i["opciones"].items():
        print(f"{letra} {opcion}")
    
    respuesta=input("Escribe tu respuesta aquí").upper()
    while respuesta not in ["A","B", "C", "D"]:
        respuesta=input("Respuesta no válida, elige una opción de la A a la D").upper()


    if respuesta == i["respuesta_correcta"]:
        print("¡Respuesta correcta!\n")
        puntuacion+=1

    else:
        if vidas > 1:    
            print(f"Respuesta incorrecta, la respuesta correcta era {i['respuesta_correcta']}.\nTe quedan {vidas} vidas.\n")
            vidas-=1
        elif vidas == 1:
            print((f"Respuesta incorrecta, la respuesta correcta era {i['respuesta_correcta']}.\nTe queda {vidas} vida.\n"))
            vidas-=1    

        else:
            print("Has cometido 3 errores. ¡El juego ha terminado!")
            break

    contador+=1

print(f"¡Juego terminado, {nombre}!\n¡Enhorabuena, has ganado!\nTu puntuación fue: {puntuacion}/{num_preguntas}")
```
